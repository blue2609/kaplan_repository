```python
response = requests.get('https://downloads.tableau.com/esdalt/2020.3.8/TableauDesktop-64bit-2020-3-8.exe',allow_redirects=True)
open('TableauDesktop.exe','wb').write(response.content)
```

```python
import shlex
import subprocess 

packageID = 'Tableau-Desktop'
commandList = shlex.split(f'choco search {packageID} --all --local-only')
# commandList = shlex.split(f'choco search {packageID} --all --verbose')
process = subprocess.Popen(commandList,stdout=subprocess.PIPE)
output = process.communicate()[0]
print(output)
```
